```python
data = pd.read_csv("/content/drive/MyDrive/data/pokemon.csv")
```

#**MATPLOTLIB**

*   Görselleştirmeye yarayan bir kütüpanedir.
*   En temel ve en kolay plotlar line, scatter ve histogram plotlardır.
*   **Line plot** incelenen özelliği zaman (gün, hafta, ay, yıllar vb) içinde gösteren grafiklerdir. x ekseni zaman olarak alınır.
*   **Scatter plot** iki feature'u karşılaştırmak için kullanılır.
*   **Histogram** sayısal verilerin dağılımını anlamak için kullanılır.
*   Dağılım feature sıklığı demektir.







```python
#LINE PLOT

data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

Line plot için kullanılan parametler;

*   **kind** parametresi ile hangi plot türünün çizdirileceği belirlenir.
*   **label** parametresi birden fazla sütun aynı anda çizdiriliyorsa seçilen sütunun adını belirtir.
*   **legend** label adını yazdırmak için kullanılır.
*   **x label/y label** x ve y labellarına isim verilir.
*   **color** parametresi ile renk seçilir. Birden fazla sütun seçildiğinde, sütunların karışmaması için aynı renk seçilmemesi önemlidir.
*   **linewidth** parametresi çizgilerin kalınlığını belirtir.
*   **alpha** parametresi çizgilere saydamlık kazandırır.
*   **grid = True** parametresi ile grafik çizilirken x eksenindeki veriler ile y eksenindeki veriler birleştirilir. Daha iyi yorumlama sağlar. 
*   **linestyle** ile grafik çizdirilirken çizgilerin türü belirlenir. Eğer kulllanılmazsa default olarak '-' çizdirilir. '-', ':', '-.', '--' türleri vardır.











```python
# Scatter Plot 
# x = attack, y = defense
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot') 
plt.show()
```

Scatter plot iki feature arasındaki korelasyonu anlamak için kullanılır. Bu grafikte Defense ve Attack sütunları seçilmiştir. Burada pozitif yönlü orta seviyeli bir korelasyon olduğu söylenebilir. Corelation Map'e bakıldığında aralarındaki ilişki 0.4386872'tir. Bu veri de grafikten yapılan yorumu destekler niteliktedir.

Scatter plot çizdirilirken x ve y eksenlerine seçilen featurelar yazılmalıdır.




```python
# Histogram
# bins = number of bar in figure
data.Speed.plot(kind = 'hist',bins = 50,figsize = (5,5))
plt.show()
```

Histogramda;
*   **bins** parametresi kullanılmazsa default olarak 10 atanır. Bu örnekte 50 kutu çizdirilmiştir.
*   **x** ekseni Speed sütununu **y** ekseni frekansı/sıklığı gösterir.


```python
# clf() = cleans it up again you can start a fresh
data.Speed.plot(kind = 'hist',bins = 50)
plt.clf()
# We cannot see plot due to clf()
```

**clf** methodu çizdirilen plotların temizlenmesini sağlar.
